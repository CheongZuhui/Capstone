## Let's explore the 1st dataset involving total air traffic

```python
passengers_df.tail()
```

```python
p_df = passengers_df.set_index("Date")
p_df.head()
```

```python
p_df.info()
```

```python
# Filter dataset to start from 1989 so that it reflects only values from Changi Airport only.
p1989_df = p_df.loc['1989-01-01':]
p1989_df
```

```python
# Permanently sort the data in chronologically order
p1989_df.sort_values("Date", inplace=True)
p1989_df
```

```python
# Overall trend of total passengers
plt.figure(figsize=(10,6))
p1989_df.total_passengers.plot()
plt.title("Monthly number of passengers handled by Changi Airport from Jan 1989 to Feb 2025")
plt.show()
```

#### Findings:
1. There is an overall increase in number of air passengers handled by Changi Airport through the years from 1989 till 2025.
2. We see several dips over the years, with more drastic decrease observed around year 2003 and 2020. This is namely due to significant events experienced worldwide that could have impacted the travel/aviation industry:\
   (a) Year 2003-2004 - SARS outbreak\
   (b) Year 2020-2022 - COVID-19
3. Post COVID-19 period saw steep increase in number of air passengers as borders open up and air travel resumes.

```python
# Decomposing the data to look at higher granularity (quarter, year) to smoooth out the noise.
fig, ax = plt.subplots(1, 2, figsize=(18, 4))

p1989_df.resample("QS")["total_passengers"].mean().plot(ax=ax[0], 
                                                        title="Quarterly avg no. of passengers handled by Changi Airport from Jan 1989 to Feb 2025")

p1989_df.resample("YS")["total_passengers"].mean().plot(ax=ax[1], 
                                                        title="Yearly avg no. of passengers handled by Changi Airport from 1989 to Feb 2025")
plt.show()
```

Overall trend observed when data is grouped to quarters and years is generally very similar to directly plotting the figures by month.

```python
# Plotting monthly passengers arrivals, departure and transit.
plt.figure(figsize=(10,6))
plt.plot(p1989_df['total_passengers_arr'], label = 'Arrivals')
plt.plot(p1989_df['total_passengers_dep'], label = 'Departures')
plt.plot(p1989_df['total_passengers_transit'], label = 'Transit')
plt.legend()
plt.title ("Monthly no. of passengers arriving at, departing from and transiting through Changi Airport")
plt.show()
```

#### Findings:
1. Trend lines for total passengers arrival and departure are very similar, with very close numbers.
2. Total passengers arrival and departure trend line also looks very close to what is observed when we plot total passengers over time, suggesting that arrivals and departures are the main contributor to total air traffic in Singapore.
3. This is opposed to the monthly number of passengers transiting through Changi Airport. The figures remained more or less the same over the years since 1989 till 2025, with slight decrease in recent years due to COVID-19 (around Year 2020).

```python
# Let's look at average total number of passengers handled by Changi Airport, grouped by month.
totalp_month_df = p1989_df.groupby(['month'])['total_passengers'].agg(['mean'])

plt.figure(figsize=(10,6))
totalp_month_df.plot.bar()
plt.title("Avg no. of air passengers handled by Changi Airport, grouped by month")
```

#### Findings:
1. The number of air passengers handled by Changi Airport is higher towards the end of the year in December, crossing into month of January. This coincides with end-of-year holiday period around the world (e.g. Christmas/New Year holidays) as well as winter breaks where travels may be more frequent.
2. We also see more air passengers during the months of June to August. This again coincides with summer breaks around the world, where we can expect more people to travel.

```python
# Seasonal decomposition of total air passengers
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
# Extract total passengers column only --> apply decomposition
decom_p_df = p1989_df['total_passengers']

decomposition = seasonal_decompose(decom_p_df)

fig = plt.figure()  
fig = decomposition.plot()  
fig.set_size_inches(12, 6)
```

#### Findings:
1. Trend: Air traffic at Changi Airport showed upward trend, increasing over the years except for the dip in 2020 due to COVID-19 travel restrictions.
2. Seasonality: Recurring pattern on a yearly cycle, with higher numbers towards the end of the year.

## Let's explore the breakdown of arrivals & departures by different regions (dataframe #2).

```python
regions_df.head()
```

```python
# We will standardise the timeframe to Jan 1989 onwards so it only reflects Changi Airport's data.

r_df = regions_df.set_index("date")
reg1989_df = r_df.loc['1989-01-01':]
reg1989_df
```

```python
# Noting that there are values 'na' for data that is not available --> will replace na with null so that we could use pandas to handle these null cells.
import numpy as np
reg1989_df = reg1989_df.replace(to_replace='na', value=np.nan)
reg1989_df
```

```python
# Check number of null values there are in each column
reg1989_df.isna().sum()
```

```python
# Null values are mainly present in 'other regions' column --> we will drop the column from analysis and focus on the remaining specified regions -
# South East Asia, North East Asia, South Asia, Middle East, Oceania, Europe and North America
reg1989_df = reg1989_df.drop(columns=['other_regions_total_dep', 'other_regions_total_arr'])
reg1989_df.info()
```

```python
# Now that we have no more null values in the columns, we will convert all datatypes from string to int.
reg1989_df = reg1989_df.astype({'sea_total_dep':int, 
                   'nth_east_asia_total_dep':int, 
                   'sth_asia_total_dep':int,
                   'mid_east_total_dep':int,
                   'oceania_total_dep':int,
                   'europe_total_dep':int,
                   'nth_america_total_dep':int,
                   'sea_total_arr':int, 
                   'nth_east_asia_total_arr':int, 
                   'sth_asia_total_arr':int,
                   'mid_east_total_arr':int,
                   'oceania_total_arr':int,
                   'europe_total_arr':int,
                   'nth_america_total_arr':int}
                   )
reg1989_df.info()
```

```python
reg1989_df.head()
```

```python
# Explore monthly departure by regions columns

plt.figure(figsize=(10,6))
plt.plot(reg1989_df['sea_total_dep'], label='South East Asia')
plt.plot(reg1989_df['nth_east_asia_total_dep'], label='North East Asia')
plt.plot(reg1989_df['sth_asia_total_dep'], label='South Asia')
plt.plot(reg1989_df['mid_east_total_dep'], label='Middle East')
plt.plot(reg1989_df['oceania_total_dep'], label='Oceania')
plt.plot(reg1989_df['europe_total_dep'], label='Europe')
plt.plot(reg1989_df['nth_america_total_dep'], label='North America')
plt.legend()
plt.title("Monthly no. of passengers DEPARTING from Changi Airport by Regions, Jan 1989 to Feb 2025")
plt.show()
```

#### Findings: 
1. In general, there is an increasing trend in no. of departing passengers handled by Changi Airport over the years.
2. Similar to what we observe in total air passenger load from dataset #1, there were 2 major dips - one around 2003 and one around 2020.

```python
# Explore monthly arrivals by regions columns

plt.figure(figsize=(10,6))
plt.plot(reg1989_df['sea_total_arr'], label='South East Asia')
plt.plot(reg1989_df['nth_east_asia_total_arr'], label='North East Asia')
plt.plot(reg1989_df['sth_asia_total_arr'], label='South Asia')
plt.plot(reg1989_df['mid_east_total_arr'], label='Middle East')
plt.plot(reg1989_df['oceania_total_arr'], label='Oceania')
plt.plot(reg1989_df['europe_total_arr'], label='Europe')
plt.plot(reg1989_df['nth_america_total_arr'], label='North America')
plt.legend()
plt.title("Monthly no. of passengers ARRIVING into Changi Airport by Regions, Jan 1989 to Feb 2025")
plt.show()
```

#### Findings: 
1. In general, there is an increasing trend in no. of arriving passengers handled by Changi Airport over the years, almost identical to plot of departing passengers by regions. This makes sense as the 2 numbers are tied to how well-connected Singapore is to that particular region e.g. number of flights flying in and out of the region to Singapore).
2. Similar to what we observe in total air passenger load from dataset #1, there were 2 major dips - one around 2003 and one around 2020.

```python
# We will now combine departures and arrivals to get the total passenger load from each regions. Note that the numbers does not include any transit no.
reg1989_df['sea_total']= reg1989_df['sea_total_arr'] + reg1989_df['sea_total_dep']
reg1989_df['nth_east_asia_total'] = reg1989_df['nth_east_asia_total_arr'] + reg1989_df['nth_east_asia_total_dep']
reg1989_df['sth_asia_total'] = reg1989_df['sth_asia_total_arr'] + reg1989_df['sth_asia_total_dep']
reg1989_df['mid_east_total'] = reg1989_df['mid_east_total_arr'] + reg1989_df['mid_east_total_dep']
reg1989_df['oceania_total'] = reg1989_df['oceania_total_arr'] + reg1989_df['oceania_total_dep']  
reg1989_df['europe_total'] = reg1989_df['europe_total_arr'] + reg1989_df['europe_total_dep']    
reg1989_df['nth_america_total'] = reg1989_df['nth_america_total_arr'] + reg1989_df['nth_america_total_dep']

reg1989_df
```

```python
# Creating a new dataframe just to store the columns containing total departures and arrivals from each region.
region_total_df = reg1989_df[["sea_total", "nth_east_asia_total", "sth_asia_total", "mid_east_total",
                             "oceania_total", "europe_total", "nth_america_total"]]

region_total_df.head()
```

```python
region_total_df.mean().sort_values(ascending=False)
```

```python
region_total_df.mean().sort_values(ascending=False).plot.bar()
plt.title("Average no. of passengers from each region handled by Changi Airport from Jan 1989 to Feb 2025")
```

#### Findings:
- Average no. of passengers handled by Changi Airport since Jan 1989 is the highest for South East Asia at about 1.21 million. This is followed by North East Asia (approx 0.71 million; consisting of countries such as China, Japan) and Oceania (approx 0.33 million; consisting of countries such as Australia, New Zealand).
- This is within expectations as the number of flights connecting Singapore to the rest of South East Asia is probably the highest given the proximity. For e.g. SG-KL route was named world's busiest international air link back in 2019. Hence, it is logical that we see higher passenger load from South East Asia vs the other parts of the world.

```python
# Plot by monthly total passengers per region. 
plt.figure(figsize=(10,6))
plt.plot(region_total_df['sea_total'], label='South East Asia')
plt.plot(region_total_df['nth_east_asia_total'], label='North East Asia')
plt.plot(region_total_df['sth_asia_total'], label='South Asia')
plt.plot(region_total_df['mid_east_total'], label='Middle East')
plt.plot(region_total_df['oceania_total'], label='Oceania')
plt.plot(region_total_df['europe_total'], label='Europe')
plt.plot(region_total_df['nth_america_total'], label='North America')
plt.legend()
plt.title("Total monthly no. of passengers arriving & departing into Changi Airport by Regions, Jan 1989 to Feb 2025")
plt.show()
```

```python
# Resample to see average figure by the quarter and year.
plt.figure(figsize=(10,6))
region_total_df.resample("QS")['sea_total'].mean().plot(label='South East Asia')
region_total_df.resample("QS")['nth_east_asia_total'].mean().plot(label='North East Asia')
region_total_df.resample("QS")['sth_asia_total'].mean().plot(label='South Asia')
region_total_df.resample("QS")['mid_east_total'].mean().plot(label='Middle East')
region_total_df.resample("QS")['oceania_total'].mean().plot(label='Oceania')
region_total_df.resample("QS")['europe_total'].mean().plot(label='Europe')
region_total_df.resample("QS")['nth_america_total'].mean().plot(label='North America')
plt.title("Quarterly avg no. of passengers handled by Changi Airport by Regions from Jan 1989 to Feb 2025")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
region_total_df.resample("YS")['sea_total'].mean().plot(label='South East Asia')
region_total_df.resample("YS")['nth_east_asia_total'].mean().plot(label='North East Asia')
region_total_df.resample("YS")['sth_asia_total'].mean().plot(label='South Asia')
region_total_df.resample("YS")['mid_east_total'].mean().plot(label='Middle East')
region_total_df.resample("YS")['oceania_total'].mean().plot(label='Oceania')
region_total_df.resample("YS")['europe_total'].mean().plot(label='Europe')
region_total_df.resample("YS")['nth_america_total'].mean().plot(label='North America')
plt.title("Yearly avg no. of passengers handled by Changi Airport by Regions from Jan 1989 to Feb 2025")
plt.legend()
plt.show()
```

#### Findings:
Looking at the yearly average graph gives us better visualisation of the trends with less noise.
1. Overall, there is an upward trend across all regions in terms of total passengers departing from and arriving at Changi Airport. This corresponds to the combined total numbers trend we saw during EDA of dataset #1.
2. South East Asia has been the top region in terms of passenger load handled by Changi Airport, even after COVID-19. This is followed by North East Asia and Oceania.
3. At around Year 2012, average no. of passenger from Europe was overtaken by South Asia, suggesting a rapidly growing market that Singapore could explore if we wish to remain competitive.
4. No. of departing and arriving passengers from North America and Middle East remained consistently low throughout the years when compared to other regions. The increment (i.e. slope) is also generally mild. This is likely due to lack of demand and low number of connecting flights to these regions. We should also note that the number of countries included within these regions is likely lower than that in other regions as well (e.g. SEA or North East Asia).
However, these are also potential opportunities for growth that Changi Airport may wish to look into to grow passenger volume passing through the airport.
6. Post COVID-19 period saw rebound in no. of passengers from all regions, with the recovery being quickest for South East Asia. Initial recovery for North East Asia was relatively slow vs Oceania, South Asia and Europe. However, by around mid-2022, the numbers quickly picked up and overtook these few regions as major countries such as Japan, Korea and China slowly opened up their borders.

### Below are codes to "melt" the tables for figures related to Regions for Dashboard building subsequently.

```python
region_total_df
```

```python
region_melt = region_total_df.reset_index()
region_melt
```

```python
# Renaming the columns 
region_melt.rename(columns={"sea_total":"South East Asia",
                            "nth_east_asia_total":"North East Asia",
                            "sth_asia_total":"South Asia",
                            "mid_east_total":"Middle East",
                            "oceania_total":"Oceania",
                            "europe_total":"Europe",
                            "nth_america_total":"North America"}, inplace=True)

# Unpivoting the columns to facilitate with dashboard building later
region_total_melt = pd.melt(region_melt, id_vars=['date'], value_name='Total')
region_total_melt
```

```python
# Creating a new dataframe just to store the columns containing departure figures from each region.
region_depart_df = reg1989_df[['sea_total_dep', 'nth_east_asia_total_dep', 'sth_asia_total_dep', 'mid_east_total_dep',
                             'oceania_total_dep', 'europe_total_dep', 'nth_america_total_dep']]

# Repeat the unpivoting process for this departure by regions dataframe
region_depart_melt = region_depart_df.reset_index()

region_depart_melt.rename(columns={"sea_total_dep":"South East Asia",
                            "nth_east_asia_total_dep":"North East Asia",
                            "sth_asia_total_dep":"South Asia",
                            "mid_east_total_dep":"Middle East",
                            "oceania_total_dep":"Oceania",
                            "europe_total_dep":"Europe",
                            "nth_america_total_dep":"North America"}, inplace=True)

region_depart_melt = pd.melt(region_depart_melt, id_vars=['date'], value_name='Departure')
region_depart_melt
```

```python
# Creating a new dataframe just to store the columns containing arrival figures from each region.
region_arrival_df = reg1989_df[['sea_total_arr', 'nth_east_asia_total_arr', 'sth_asia_total_arr', 'mid_east_total_arr',
                             'oceania_total_arr', 'europe_total_arr', 'nth_america_total_arr']]

# Repeat the unpivoting process for this arrival by regions dataframe
region_arrival_melt = region_arrival_df.reset_index()

region_arrival_melt.rename(columns={"sea_total_arr":"South East Asia",
                            "nth_east_asia_total_arr":"North East Asia",
                            "sth_asia_total_arr":"South Asia",
                            "mid_east_total_arr":"Middle East",
                            "oceania_total_arr":"Oceania",
                            "europe_total_arr":"Europe",
                            "nth_america_total_arr":"North America"}, inplace=True)

region_arrival_melt = pd.melt(region_arrival_melt, id_vars=['date'], value_name='Arrival')
region_arrival_melt
```

```python
# Combining the 3 unpivoted table to give a master table containing total, departure and arrival numbers by Region
region_melt_all = pd.concat([region_total_melt, region_depart_melt['Departure'], region_arrival_melt['Arrival']], axis=1)
region_melt_all
```

```python
# Export to csv to be used in Power BI
region_melt_all.to_csv("region_melt_all.csv")
```
