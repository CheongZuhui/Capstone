```python
import sqlite3
import pandas as pd
```

```python
con = sqlite3.connect('air_traffic_db.db')
cursor = con.cursor()
```

```python
air_traffic_df = pd.read_csv('air_traffic_total.csv')
print(air_traffic_df.shape)
air_traffic_df.head()
```

```python
air_traffic_df.info()
```

```python
air_traffic_df.rename(columns={'Total Passengers' : 'total_passengers', 
                               'Total Passengers Arriving' : 'total_passengers_arr',
                               'Total Passengers Departing' : 'total_passengers_dep',
                               'Total Passengers In Transit': 'total_passengers_transit'}, inplace = True)
air_traffic_df.head()
```

```python
# storing the df in the air_traffic_db db
air_traffic_df.to_sql(name='air_traffic_cln',con=con, index=False, if_exists='replace')
```

```python
sql_query = "SELECT Date, total_passengers, total_passengers_arr, total_passengers_dep, total_passengers_transit FROM air_traffic_cln"
# Selecting only columns of interest i.e. passengers movement
```

```python
df_passengers = pd.read_sql(sql_query,con)
df_passengers
```

```python
# Change datatype for Date to datetime
df_passengers["Date"] = pd.to_datetime(df_passengers["Date"], yearfirst=True)
df_passengers.head()
```

```python
df_passengers.info()
```

```python
# Loading in dataset for arrivals and departure based on regions/country
dep_region_df = pd.read_csv('air_dep_region.csv')
print(dep_region_df.shape)
dep_region_df.head()
```

```python
# storing the df in the air_traffic_db db
dep_region_df.to_sql(name='dep_region_cln',con=con, index=False, if_exists='replace')
```

```python
dep_region_df.columns
```

```python
# Selecting only columns of interest i.e. total passengers movement in different regions
sql_query = """SELECT date, total_passengers_dep, sea_total_dep, nth_east_asia_total_dep, sth_asia_total_dep, 
mid_east_total_dep, oceania_total_dep, europe_total_dep, nth_america_total_dep, other_regions_total_dep 
FROM dep_region_cln"""
```

```python
df_dep_reg = pd.read_sql(sql_query,con)
df_dep_reg
```

```python
arr_region_df = pd.read_csv('air_arr_region.csv')
print(arr_region_df.shape)
arr_region_df.head()
```

```python
# storing the df in the air_traffic_db db
arr_region_df.to_sql(name='arr_region_cln',con=con, index=False, if_exists='replace')
```

```python
arr_region_df.columns
```

```python
# Selecting only columns of interest i.e. total passengers movement in different regions
sql_query = """SELECT date, total_passengers_arr, sea_total_arr, nth_east_asia_total_arr, sth_asia_total_arr, 
mid_east_total_arr, oceania_total_arr, europe_total_arr, nth_america_total_arr, other_regions_total_arr 
FROM arr_region_cln"""
```

```python
df_arr_reg = pd.read_sql(sql_query,con)
df_arr_reg
```

```python
# Store the 2 df to SQL database
df_dep_reg.to_sql(name='dep_by_region',con=con, index=False, if_exists='replace')
df_arr_reg.to_sql(name='arr_by_region',con=con, index=False, if_exists='replace')

# Join the 2 tables above for departure by regions and arrivals by regions into 1 table so that we could analyse them in totality later
sql_query = """SELECT dep_by_region.date, total_passengers_dep, sea_total_dep, nth_east_asia_total_dep, sth_asia_total_dep, 
mid_east_total_dep, oceania_total_dep, europe_total_dep, nth_america_total_dep, other_regions_total_dep, 
total_passengers_arr, sea_total_arr, nth_east_asia_total_arr, sth_asia_total_arr, 
mid_east_total_arr, oceania_total_arr, europe_total_arr, nth_america_total_arr, other_regions_total_arr
FROM dep_by_region
join arr_by_region on dep_by_region.date = arr_by_region.date"""
df_regions = pd.read_sql(sql_query,con)
df_regions
```

```python
df_regions.info()
```

```python
# Changing datatype of date to datetime
df_regions['date'] = pd.to_datetime(df_regions['date'])
```

```python
df_regions.info()
```

```python
# Hence, now we have 2 tables/df: 
# (1) df_passengers = total air traffic over the years, 
# (2) df_regions = no. of passengers arriving from and departing to a particular region
df_passengers.head()
```
